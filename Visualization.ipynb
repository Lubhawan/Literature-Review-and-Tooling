```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('dat.csv')
```

```python
df
```

```python
tot_data = [df.iloc[2][2:],df.iloc[9][2:],df.iloc[12][2:],df.iloc[17][2:]]
fig = plt.figure(figsize =(10, 7),dpi=150)
 
# Creating axes instance

ax = fig.add_axes([0, 0, 1, 1])
# Creating plot
bp = ax.boxplot(tot_data)

ax.set_xticklabels(['Total Passenger Vehicles (PVs)', 'Total Commercial Vehicles (CVs)', 'Total Three Wheelers', 'Total Two wheelers'])
plt.ylabel('Total Production (NUmbers)')
plt.title('Production numbers of different type of vehicles')
# show plot
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=75)
plt.bar(df.columns[2:],df.iloc[2][2:])
# ax.barh(df.columns[2:],df.iloc[2][2:])
# plt.text(df.columns[2:],df.iloc[2][2:])
plt.bar(df.columns[2:],df.iloc[9][2:])
# ax.barh(df.columns[2:],df.iloc[2][2:])
# plt.text(df.columns[2:],df.iloc[9][2:])
ax = plt.gca()
ax.set_xticklabels(df.columns[2:], rotation = 90)
plt.legend(['Total Passenger Vehicles (PVs)','Total Commercial Vehicles (CVs)'])
plt.xlabel('Year')
plt.ylabel('Total Production (NUmbers)')
```
